```python
import numpy as np
import pandas as pd
```

```python
jobs = pd.read_csv("DataAnalyse.csv")
```

```python
jobs["Revenue"].unique()
```

```python
jobs["Revenue"].value_counts().plot.pie(figsize=(10, 8), autopct="%1.1f%%")
```

```python
jobs["Salary Estimate"].value_counts().plot.pie(figsize=(12, 12), autopct="%1.1f%%")
```

```python
jobs["Salary Estimate"].value_counts()
```

```python
jobs = jobs[jobs["Salary Estimate"].notnull()]
```

```python
jobs["Bottom Salary Estimate($K)"] = jobs["Salary Estimate"].str.extract("(\d+)").astype(int)
```

```python
jobs["Top Salary Estimate($K)"] = jobs["Salary Estimate"].str.extract("-\$(\d+)").astype(float)
```

```python
jobs.head()
```

```python
pd.cut(jobs["Bottom Salary Estimate($K)"], np.arange(0, 120, 10))
```

```python
jobs["Bottom Salary Estimate($K)"].groupby(pd.cut(jobs["Bottom Salary Estimate($K)"], np.arange(0, jobs["Bottom Salary Estimate($K)"].max() + 10, 10))).count().plot.pie(figsize=(10, 8), autopct="%1.1f%%")
```

```python
jobs["Top Salary Estimate($K)"].groupby(pd.cut(jobs["Top Salary Estimate($K)"], np.arange(0, jobs["Top Salary Estimate($K)"].max() + 10, 10))).count().plot.pie(figsize=(10, 8))
```

```python
jobs["Mean Salary Estimate($K)"] = (jobs["Bottom Salary Estimate($K)"] + jobs["Top Salary Estimate($K)"]) / 2
```

```python
jobs.head()
```

```python
jobs["Mean Salary Estimate($K)"].groupby(pd.cut(jobs["Mean Salary Estimate($K)"], np.arange(0, jobs["Mean Salary Estimate($K)"].max() + 10, 10))).count().plot.pie(figsize=(10, 8), autopct="%1.1f%%")
```

```python
jobs["Mean Salary Estimate($K)"].groupby(pd.cut(jobs["Mean Salary Estimate($K)"], np.arange(0, jobs["Mean Salary Estimate($K)"].max() + 10, 10))).count().plot.bar(figsize=(14, 6))
```
